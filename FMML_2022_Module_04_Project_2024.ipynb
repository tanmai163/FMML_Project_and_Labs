## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the functions
def f1(x, y):
    return 1 / (3**(-x**2 - y**2) + 1)

def f2(x, y):
    r = 3**(-x**2 - y**2)
    return 1 / (r + 1)

# Define the partial derivatives (gradients) of the functions
def df1_dx(x, y):
    return 2 * x * np.log(3) * f1(x, y) / (3**(-x**2 - y**2) + 1)**2

def df1_dy(x, y):
    return 2 * y * np.log(3) * f1(x, y) / (3**(-x**2 - y**2) + 1)**2

def df2_dx(x, y):
    r = 3**(-x**2 - y**2)
    return 2 * x * r * f2(x, y) / (r + 1)**2

def df2_dy(x, y):
    r = 3**(-x**2 - y**2)
    return 2 * y * r * f2(x, y) / (r + 1)**2

# Define the gradient descent function
def gradient_descent_2d(f, df_dx, df_dy, initial_guess, learning_rate=0.01, precision=0.0001, max_iters=1000):
    current_guess = np.array(initial_guess)
    trajectory = [current_guess]

    for _ in range(max_iters):
        gradient_x = df_dx(current_guess[0], current_guess[1])
        gradient_y = df_dy(current_guess[0], current_guess[1])
        gradient = np.array([gradient_x, gradient_y])

        current_guess = current_guess - learning_rate * gradient
        trajectory.append(current_guess)

        if np.linalg.norm(gradient) < precision:
            break

    return current_guess, np.array(trajectory)

# Initial guess for gradient descent
initial_guess_f1 = [0.0, 0.0]
initial_guess_f2 = [0.0, 0.0]

# Calling gradient descent function for f1
min_point_f1, trajectory_f1 = gradient_descent_2d(f1, df1_dx, df1_dy, initial_guess_f1)

# Calling gradient descent function for f2
min_point_f2, trajectory_f2 = gradient_descent_2d(f2, df2_dx, df2_dy, initial_guess_f2)

```

## Batch Gradient Descent with SymPy

```python
import numpy as np
from sympy import symbols, diff

a, b = symbols('x y')


f = a**2 + b**2


multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])

for n in range(max_iter):
    gradient_x = diff(f, a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = diff(f, b).evalf(subs={a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplier * gradients

# Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f.evalf(subs={a: params[0], b: params[1]}))
```

```python
nr_thetas = 200
th_0 = np.linspace(start=-1, stop=3, num=nr_thetas)
th_1 = np.linspace(start=-1, stop=3, num=nr_thetas)
plot_t0, plot_t1 = np.meshgrid(th_0, th_1)
```

### Calc MSE using nested for loops

```python
def mse(y, y_hat):
    mse_calc = np.mean((y - y_hat)**2, axis=0)
    return mse_calc
```

```python
# Function to calculate MSE
def mse(y, y_hat):
    return np.mean((y - y_hat)**2)

# Partial derivatives of MSE with respect to theta_0 and theta_1
def mse_partial_theta0(y, y_hat):
    return -2 * np.mean(y - y_hat)

def mse_partial_theta1(y, y_hat, x):
    return -2 * np.mean((y - y_hat) * x)

# Initialize arrays
plot_cost = np.zeros((nr_thetas, nr_thetas))
partial_theta_0 = np.zeros((nr_thetas, nr_thetas))
partial_theta_1 = np.zeros((nr_thetas, nr_thetas))

# Loop to calculate MSE and partial derivatives
for i in range(nr_thetas):
    for j in range(nr_thetas):
        theta_0 = plot_t0[i][j]
        theta_1 = plot_t1[i][j]
        y_hat = theta_0 + theta_1 * x_5

        # Calculate MSE
        plot_cost[i][j] = mse(y_5, y_hat)

        # Calculate partial derivatives
        partial_theta_0[i][j] = mse_partial_theta0(y_5, y_hat)
        partial_theta_1[i][j] = mse_partial_theta1(y_5, y_hat, x_5)

print('Shape of plot_t0', plot_t0.shape)
print('Shape of plot_t1', plot_t1.shape)
print('Shape of plot_cost', plot_cost.shape)
print('Partial derivative with respect to theta_0:\n', partial_theta_0)
print('Partial derivative with respect to theta_1:\n', partial_theta_1)
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```
